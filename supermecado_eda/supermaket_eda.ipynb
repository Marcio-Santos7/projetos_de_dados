```python
# O método info() é uma ferramenta essencial para inspeção detalhada da estrutura de um DataFrame pandas. Ele fornece um resumo conciso mas completo 
# sobre o conjunto de dados:
# 1. RangeIndex: Índice do DataFrame -> Mostra o intervalo de índices (0 a N-1);
# 2. Non-Null Count: Contagem de valores não-nulos -> Indica quantos valores válidos existem em cada coluna;
# 3. Dtype: Tipo de dado:
#    3.1. object: Strings/texto;
#    3.2. int64: números inteiros;
#    3.3. float64: Números decimais;
#    3.4. bool: Valores booleanos;
#    3.5. datetime64: Datas.
```

```python
df.info()
```

```python
# O método describe() é uma ferramenta poderosa para análise estatística descritiva de DataFrames e Series pandas. Ele gera automaticamente estatísticas
# resumidas que ajudam a entender a distribuição e características dos dados.
# 1. count: Número de valores não-nulos;
# 2. mean: Média aritmética;
# 3. std: Desvio padrão;
# 4. min: Valor mínimo;
# 25%: Primeiro quartil (percentil 25);
# 50%: Mediana (segundo quartil);
# 75%: Terceiro quartil (percentil 75);
# max: Valor máximo.
```

```python
df.describe()
```

```python
# O método describe com o parâmetro (include="O") é uma variação do método describe() do pandas que se concentra especificamente em colunas com dados 
# textuais (objetos) em um DataFrame.
# Quando usado com o include="O", o método describe() fornece estatísticas descritivas diferentes das numéricas,específicas para colunas de texto:
# 1. count: Número de valores não-nulos;
# 2. unique: Quantidade de valores únicos/distintos;
# 3. top: Valor mais frequente (moda);
# 4. freq: Frequência do valor mais comum.
```

```python
df.describe(include="O")
```

```python
# O método duplicated() é usado para identificar linhas duplicadas em um DataFrame ou Series pandas. Ele retorna uma série booleana onde True indica que a
# linha é duplicada (aparece mais de uma vez) e False indica que a linha é única. Em conjunto com o método sum() informa o número total de linhas repetidas.
```

```python
df.duplicated().sum()
```

```python
# O método isnull() é usado para identificar valores ausentes ou nulos (missing values) em um DataFrame ou Series pandas. Ele retorna um objeto do mesmo 
# tamanho onde True indica que o valor é nulo e False indica que o valor é válido. Em conjunto com o método sum() mostra a contagem de nulos por coluna.
```

```python
df.isnull().sum()
```

```python
# Realizando uma filtragem que seleciona apenas as linhas onde a coluna 'Postal Code' contém valores nulos/ausentes. O resultado é um novo DataFrame 
# contendo apenas os registros problemáticos (sem código postal).
```

```python
df[df['Postal Code'].isnull()]
```

```python
# Uma forma de resolver o problema com valores ausentes é utilizando o método fillna() que preencherá os campos selecionados do DataFrame com o valor 
# especificado entre parênteses, que nesse caso será o valor 5401.
```

```python
df['Postal Code'] = df['Postal Code'].fillna(5401)
df
```

```python
# Conferindo com o método isnull().sum() que não existe mais nenhum valor nulo/ausente no DataFrame.
```

```python
df.isnull().sum()
```
